### Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/MD_Crime_Data.csv')
df=df.fillna(0)
```

```python
print(df['JURISDICTION'].unique())
print(df['YEAR'].unique())
#All counties are present.
#Years range from 1975 - 2020
```

```python
df.columns
```

```python
sns.lineplot(data=df,x='YEAR',y='GRAND TOTAL')
plt.ylabel('Total Crimes')
plt.xlabel('Year')
plt.title('Crimes committed by Year')

#Maryland's government has been doing a great job at reducing crimes commited since the mid 90s.
```

```python
total_property = df['PROPERTY CRIME TOTALS'].sum()
total_violent = df['VIOLENT CRIME TOTAL'].sum()

labels = ['Property Crime', 'Violent Crime']
sizes = [total_property, total_violent]
colors = ['skyblue', 'lightcoral']
explode = (0, 0.1)  

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title('Proportion of Property Crimes vs Violent Crimes since 1975')
plt.axis('equal')
#Since the 1975, around 85% of all crimes are property related.
```

```python
trended_crime_df = df[['YEAR','PROPERTY CRIME TOTALS','VIOLENT CRIME TOTAL']]
df2 = trended_crime_df.pivot_table(index='YEAR',values=['PROPERTY CRIME TOTALS','VIOLENT CRIME TOTAL'])
fig, ax = plt.subplots(figsize=(12,8))
ax.bar(df2.index, df2['PROPERTY CRIME TOTALS'], label='Property Crimes')
ax.bar(df2.index, df2['VIOLENT CRIME TOTAL'], label='Violent Crimes')
ax.legend()
#There has been a steady decline in Property related crimes.
```

```python
trended_crime_df2 = df[['YEAR','PROPERTY CRIME PERCENT','VIOLENT CRIME PERCENT']]
df2 = trended_crime_df2.pivot_table(index='YEAR',values=['PROPERTY CRIME PERCENT','VIOLENT CRIME PERCENT'])
fig, ax = plt.subplots(figsize=(12,8))
ax.bar(df2.index, df2['PROPERTY CRIME PERCENT'], label='Property Crimes %')
ax.bar(df2.index, df2['VIOLENT CRIME PERCENT'], label='Violent Crimes %')
ax.legend(loc='center right')
#Recently violent crimes have been making up a higher % of total crimes.
```

```python
columns_ = ['MURDER', 'RAPE', 'ROBBERY',
       'AGG. ASSAULT', 'B & E', 'LARCENY THEFT', 'M/V THEFT', 'GRAND TOTAL']

pivot_df = df.pivot_table(index='JURISDICTION',values=columns_,aggfunc='sum')
pivot_df= pivot_df.reset_index()
for column in columns_:
 pivot_df.sort_values(column,inplace=True,ascending=False)
 pivot_df.plot(kind='bar', x='JURISDICTION',y=column)
```
